```python
from collections import defaultdict
import pandas as pd
# Import the necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("ride_timestamps.csv") 
df.head()
```

```python
#df2 = pd.DataFrame(df.Raw_info.values.reshape(-1, 5), 
                #   columns=['requested_at', 'accepted_at', 'arrived_at','picked_up_at','dropped_off_at'])
df_ride_deets= df.pivot(index='ride_id',columns='event',values='timestamp').reset_index()
df_ride_deets
```

```python
#pd.to_datetime(df_ride_deets.drop('ride_id', axis=1))
#df_ride_deets.head()
#requested_at -> accepted_at -> arrived_at -> picked_up_at -> dropped_off_at
from datetime import datetime
df_ride_deets['accepted_at'] = pd.to_datetime(df_ride_deets['accepted_at'])
df_ride_deets['arrived_at'] = pd.to_datetime(df_ride_deets['arrived_at'])
df_ride_deets['dropped_off_at'] = pd.to_datetime(df_ride_deets['dropped_off_at'])
df_ride_deets['picked_up_at'] = pd.to_datetime(df_ride_deets['picked_up_at'])
df_ride_deets['requested_at'] = pd.to_datetime(df_ride_deets['requested_at'])
#df_ride_deets
#(h,m,s)= pd.to_datetime((df_ride_deets['accepted_at']- df_ride_deets['requested_at'])).split(':')
#df_ride_deets['accepted-requested'] = int(h) * 3600 + int(m) * 60 + int(s)
df_ride_deets['accepted-requested'] = (df_ride_deets['accepted_at']- df_ride_deets['requested_at']).dt.seconds
df_ride_deets['arrived-accepted'] = (df_ride_deets['arrived_at']- df_ride_deets['accepted_at']).dt.seconds
df_ride_deets['pickedup-arrived'] = (df_ride_deets['picked_up_at']- df_ride_deets['arrived_at']).dt.seconds
df_ride_deets['dropped-picked'] = (df_ride_deets['dropped_off_at']- df_ride_deets['picked_up_at']).dt.seconds

df_ride_deets.dtypes
```
